```python
import pandas as pd
```

```python
data = {
    "cars": ["camry", "civic", "optima"],
    "company" : ["toyota", 'honda', 'kia'],
    "country" : ["japan", 'japan', 'japan']
}
```

```python
df = pd.DataFrame(data=data)
```

```python
print(df)
```

```python
# 'loc' is a Pandas attribute
print(df.loc[0])
```

```python
print(df.loc[[0,1]])
```

```python
print(type(df.loc[0]))
```
