```python
import pandas as pd
```

```python
lst = [1, 2, 3, 4, 5]
```

```python
# 'Series' in pandas is a one dimensional data type. 
# Its a class. It has its own methods.

my_series = pd.Series(lst)
```

```python
print(my_series)
```

```python
# Series.add() method is adding some value to existing records.
# Ex: 1 to 1+6 =7, 2 to 2+6 = 8 etc. making entire series is addition of 6.

print(my_series.add(6))

```

```python
series_ds = pd.Series([1, 2, 3, 4, 5], index=["a", "b", "c", "d", "e"])
```

```python
print(series_ds)
```

```python
# accessing through index. 
print(series_ds["d"])
```
